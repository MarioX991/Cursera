```python
Vojvodina=final[final["Team"]=="Vojvodina"]["Score"].values
Vojvodina
```

```python
plt.figure(figsize=(20,7))
plt.title("Serbian football team %WIN\n\n(Time Period 1980-2019)\n",size=16)
plt.plot(np.arange(40),Vojvodina,label="Vojvodina",lw=3,marker='o',ms=6,mec="white",mfc="g",alpha=0.6)
plt.plot(np.arange(40),partizan,label="Partizan",c="Black",lw=4,marker='o',ms=6,mec="white",mfc="g",alpha=0.6)
plt.plot(np.arange(40),red_star,label="Red Star",c="Red",lw=4,alpha=0.6,ms=6,mec="white",mfc="g",marker='o')
plt.xticks(np.arange(0,41,10),("1980","1990","2000","2010","2019"),size=13)
plt.yticks([0,0.25,0.50,0.75,1],size=13)
plt.xlabel("Seasone",size=16)
plt.hlines(y=[0.50,0.75,0.25,0],linestyles="--",alpha=0.3,xmin=0,xmax=40)
plt.ylabel("10 Year Moving Average Win %\n",size=16)
ax= plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
plt.legend(prop={'size': 14})
plt.gray()
```
